```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# create our dataset using numpy and pandas
data = { 'id': ['T1', 'T1', 'T1', 'T1', 'T1', 'T1',
                'T2', 'T2', 'T2', 'T2', 'T2', 'T2', 
                'T3', 'T3', 'T3', 'T3',
                'T4', 'T4', 'T4', 'T4', 'T4',
                'T5', 'T5', 'T5', 'T5', 'T5', 'T5'],
        
        'item': ['Mango', 'Onion', 'Nintendo', 'Key-chain', 'Eggs', 'Yo-yo',
                'Doll', 'Onion', 'Nintendo', 'Key-chain', 'Eggs', 'Yo-yo', 
                'Mango', 'Apple', 'Key-chain', 'Eggs',
                'Mango', 'Umbrella', 'Corn', 'Key-chain', 'Yo-yo',
                'Corn', 'Onion', 'Onion', 'Key-chain', 'Ice-cream', 'Eggs'],
       }

# Calling DataFrame constructor
df = pd.DataFrame(data)
df
```

```python
# add new column with constant value 1
df['value'] = df.apply(lambda x: 1, axis=1)
```

```python
df
```

```python
# Reshape data to pivot form
df2 = pd.pivot_table(df,index='id', columns='item', values='value')
```

```python
df2
```

```python
# remove 'nan' and replace it by 0
df3 = df2.fillna(0)
```

```python
df3
```

```python
# run apriori algo
frequent_itemsets = apriori(df3, min_support=0.2, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules
```
